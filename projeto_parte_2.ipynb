**Como não há garantia de padronização entre o formato dos dados fornecidos, vamos verificar se todos os dataframes possuem a mesma quantidade de colunas e se essas são iguais**

```python
def check_dataframe_consistency(dataframes):
    # Vamos usar o primeiro dataframe (referente ao primeiro semestre de 2014) como referência
    num_cols = len(dataframes[next(iter(dataframes))].columns)
    col_names = set(dataframes[next(iter(dataframes))].columns)

    for key, df in dataframes.items():
        if len(df.columns) != num_cols:
            return f"Dataframe {key} tem uma quantidade diferente de colunas."

        if set(df.columns) != col_names:
            different_columns = set(col_names) ^ set(df.columns)
            return f"Dataframe {key} tem colunas diferentes. Segue nome da coluna destoante e o nome da mesma coluna nos outros arquivos: {different_columns}"

    return "Todos os dataframes possuem a mesma quantidade de colunas e mesmo nome para elas."

consistent = check_dataframe_consistency(dataframes)
print(consistent)
```

Todos os arquivos possuem a mesma quantidade de colunas e apenas o título de uma é diferente. Vamos igualar e verificar novamente a consistência.

```python
def equalize_column_names(dataframes):
  # Define o primeiro dataframe como referência
  reference_df = next(iter(dataframes.values()))

  for df in dataframes.values():
    df.columns = reference_df.columns

  return dataframes

dataframes = equalize_column_names(dataframes)

consistent = check_dataframe_consistency(dataframes)
print(consistent)
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Variáveis preditoras e variável alvo
X = gasolina_df[['Regiao', 'Estado', 'Municipio', 'Valor_de_Compra', 'Bandeira', 'Ano', 'Mes']]
y = gasolina_df['Valor_de_Venda']

# Colunas categóricas
categorical_features = ['Regiao', 'Estado', 'Municipio', 'Bandeira']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Modelo de CatBoost
model = CatBoostRegressor(cat_features=categorical_features,
    iterations=1000,  # Aumente o número de iterações
    depth=8,          # Aumente a profundidade das árvores
    learning_rate=0.05,  # Ajuste a taxa de aprendizado
    loss_function='RMSE',
    verbose=200)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print(f'Erro médio quadrático: {mse}')

importances = model.get_feature_importance()
features = X.columns

for i, feature in enumerate(features):
    print(f'{feature}: {importances[i]}')
```

```python
# Faça previsões no conjunto de treinamento e teste
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calcule o erro médio quadrático para treinamento e teste
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

# Calcular o R^2
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

print(f'Erro médio quadrático no treinamento: {mse_train}')
print(f"R^2 no treinamento: {r2_train}")

print(f'Erro médio quadrático no teste: {mse_test}')
print(f"R^2 no teste: {r2_test}")
```

# Seleção de hiper-parâmetros

```python
from sklearn.model_selection import GridSearchCV

param_grids = {
    "Ridge": {
        'alpha': [0.1, 1.0, 10.0]
    },
    "DecisionTreeRegressor": {
        'max_depth': [None, 5, 10, 15],
        'min_samples_split': [2, 5, 10]
    },
    "RandomForestRegressor": {
        'n_estimators': [50, 100, 150],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5, 10]
    }
}

# Inicializar os modelos
models = {
    "Ridge": Ridge(),
    "DecisionTreeRegressor": DecisionTreeRegressor(random_state=0),
    "RandomForestRegressor": RandomForestRegressor(random_state=0)
}

# Loop sobre cada modelo e seus parâmetros
for model_name, model in models.items():
    param_grid = param_grids[model_name]

    # Definir a busca de hiperparâmetros
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, n_jobs=-1, verbose=1, scoring='neg_mean_squared_error')

    # Iniciar uma nova execução
    with mlflow.start_run():
        # Ajustar o modelo com a busca de hiperparâmetros
        grid_search.fit(X_train, y_train)

        # Melhor modelo encontrado
        best_model = grid_search.best_estimator_

        # Fazer previsões
        y_train_pred = best_model.predict(X_train)
        y_test_pred = best_model.predict(X_test)

        # Avaliar o melhor modelo
        mse_train = mean_squared_error(y_train, y_train_pred)
        r2_train = r2_score(y_train, y_train_pred)
        mse_test = mean_squared_error(y_test, y_test_pred)
        r2_test = r2_score(y_test, y_test_pred)

        # Logar o melhor parâmetro
        mlflow.log_param("best_params", grid_search.best_params_)
        mlflow.log_param("best_score", grid_search.best_score_)

        # Logar as métricas
        mlflow.log_metric("mse_train", mse_train)
        mlflow.log_metric("r2_train", r2_train)
        mlflow.log_metric("mse_test", mse_test)
        mlflow.log_metric("r2_test", r2_test)

        # Logar o modelo
        mlflow.sklearn.log_model(best_model, f"model_{model_name.lower().replace(' ', '_')}")

        # Imprimindo os resultados
        print(f"Modelo: {model_name}")
        print(f"Melhores parâmetros: {grid_search.best_params_}")
        print(f"Melhor score: {grid_search.best_score_}")
        print(f"MSE no conjunto de treinamento: {mse_train}")
        print(f"R^2 no conjunto de treinamento: {r2_train}")
        print(f"MSE no conjunto de teste: {mse_test}")
        print(f"R^2 no conjunto de teste: {r2_test}")
        print()
```
